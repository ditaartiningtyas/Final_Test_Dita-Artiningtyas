```python
# FINAL PROJECT
# Dita Artiningtyas
# Kelompok 12

import requests
import pandas as pd

# 1. API request untuk mendapatkan data return API berupa JSON seperti contoh di atas dengan menggunakan Python
url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
else:
    print("Gagal mengambil data dari API")
    data = None
```

```python
# 2. Ambil value dari key ‘data’ dalam JSON, dan transformasi menjadi sebuah DataFrame.
if data:
    news_data = data.get('data', [])
    df = pd.DataFrame(news_data)
```

```python
# 3. Transformasi data kolom ‘isoDate’ menjadi tipe data datetime
if 'isoDate' in df:
    df['isoDate'] = pd.to_datetime(df['isoDate'])
```

```python
# 4. Aggregasi data berdasarkan kolom ‘creator’, serta urutkan berdasarkan jumlah berita terbanyak.
if 'creator' in df:
    creator_counts = df['creator'].value_counts()
    sorted_df = df.groupby('creator').agg({'title': 'count'}).reset_index()
    sorted_df = sorted_df.rename(columns={'title': 'jumlah_berita'})
    sorted_df = sorted_df.sort_values(by='jumlah_berita', ascending=False)
```
